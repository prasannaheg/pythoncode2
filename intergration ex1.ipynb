```python
#APPROACH ADDING ATTRIBUTES

import pandas as pd
df=pd.DataFrame({'data':['10/2/2011','11/2/2011','12/2/2011','13/2/2001'],
                 'event':['music','poetry','theatre','comedy'],
                 'cost':[10000,5000,15000,20000]
                 })
df=pd.DataFrame(df)
df
```

```python
df['discounted_price']=df.apply(lambda a: a.cost-(a.cost*0.1),axis=1)
df
```
